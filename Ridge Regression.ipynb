```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_diabetes
data=load_diabetes()
```

```python
x=data.data
y=data.target
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression,Ridge
```

```python
model1=LinearRegression()
model2=Ridge(alpha=0.01)
```

```python
model1.fit(x_train,y_train)
model2.fit(x_train,y_train)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,model1.predict(x_test))
```

```python
r2_score(y_test,model2.predict(x_test))
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.linear_model import Lasso
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
```

```python
X,y = make_regression(n_samples=100, n_features=1, n_informative=1, n_targets=1,noise=20,random_state=13)

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)

plt.scatter(X,y)

from sklearn.linear_model import LinearRegression

reg = LinearRegression()
reg.fit(X_train,y_train)
plt.plot(X_test,reg.predict(X_test),color='r')
print(reg.coef_)
print(reg.intercept_)
```

```python
alphas = [0,1,5,10,30]
plt.figure(figsize=(12,6))
plt.scatter(X,y)
for i in alphas:
    L = Lasso(alpha=i)
    L.fit(X_train,y_train)
    plt.plot(X_test,L.predict(X_test),label='alpha={}'.format(i))
plt.legend()
plt.show()
```

```python
from sklearn.linear_model import ElasticNet
```

```python
model3=ElasticNet(alpha=0.01,l1_ratio=0.8)
```

```python
model3.fit(x_train,y_train)
```

```python
r2_score(y_test,model3.predict(x_test))
```
